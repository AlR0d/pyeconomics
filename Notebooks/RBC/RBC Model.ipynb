To see what the above preferences imply about the optimal choices of household labor supply and consumption, suppose the representative agent lives for only a single period and has no initial wealth (for simplicity also assume that the houshold has only one member). These assumptions imply that the houshold budget constraint is simply $C = wl$, where $w$ is the real wage. To find optimal consumption bundle, set up the Lagrangian: 

$$\max_{C,l}\ ln\ C + b\ ln(1 - l) + \lambda(wl - C)$$

FOC and the budget constraint yield a system of 3 equation in 3 unknowns $C, l, \lambda$:

$$\frac{1}{C} - \lambda = 0$$

$$-\frac{b}{1 - l} + \lambda w = 0$$

$$C = wl$$

One can reduce the above to a 2 equation system in 2 unkowns by eliminating $\lambda$:

$$bC + wl = w$$

$$C - wl = 0$$

Or, equivalently 

$$\begin{vmatrix} b & \ w \\ 1 & -w\end{vmatrix} \begin{vmatrix}C \\ l \end{vmatrix} = \begin{vmatrix} w \\ 0 \end{vmatrix}$$

Note that this is a linear system of equations (which makes our life much easier!). Solving the system (via your favorite method for solving linear systems of equations) yields expressions for optimal choices of $C$ and $l$ as functions of prices (and model parameters):

$$C(w; b) = \frac{1}{1 + b}w$$

$$l(w; b) = \frac{1}{1 + b}$$

$$\lambda(w; b) = \frac{1 + b}{w}$$

Note that the household's labor supply choice is indepedent of the real wage, $w$.  This independence of labor supply and the real wage is a direct consequence of logarithmic utility and the assumption that houshold has no initial wealth and thus the income and substitution effects of a price change exactly offset one another.  Also, note that the household will choose to consum a fixed fraction of its lifetime net worth (which in the case is just the value of their labor endowment for a single period!).  This 'consume a fixed fraction of start of period net worth' decision rule is a direct consequence of logarithmic preferences.  The code in the cell below uses the linear equation solver from the SciPy linear algebra package to solve for the optimal consumption-labor bundle.

```python
# Choose a value for b
b=2.5

# Set prices (i.e., wages) which agents take as given when solving optimization problem
w = 10

# Define the coefficient matrix, A, and vector of dependent values, d, for our two equation system
A = np.array([[b, w], 
              [1, -w]]) 
d = np.array([[w], 
              [0]])

# Solve the system of equations and assign the optimal choices for consumption and labor
C_star, l_star = linalg.solve(A, d)[0,0], linalg.solve(A, d)[1,0] 
u_star = u(C_star, l_star)

print "Optimal choice of consumption:         ", C_star 
print "Optimal choice of labor supply:        ", l_star
print "Utility associated with optimal bundle:", u_star
```

The fact that the level of wage does not effect labor supply in a static model does not mean that variations in the real wage do not impact household labor supply when the time horizon is more than one period.  Suppose that the representative household lives for two periods and that there is no uncertainty about future prices.  Because of logarithmic preferences, the household will again follow the decision rule 'consume a fraction fixed fraction of the PDV of lifetime net worth.'  We can derive this decision rule formally as follows. First, note that the household budget constraint is 

$$C_{0} + \frac{1}{1 + r_{1}}C_{1} = w_{0}l_{0} + \frac{1}{1 + r_{1}}w_{1}l_{1}$$ 

where $r_{1}$ is the real interest rate.  The Lagrangian for the household's two period optimization problem is 

$$\max_{\{C_{t}\}, \{l_{t}\}} ln(C_{0}) + b\ ln(1-l_{0}) + \beta[ln(C_{1}) + b\ ln(1-l_{1})] + \lambda\left[w_{0}l_{0} + \frac{1}{1 + r_{1}}w_{1}l_{1} - C_{0} - \frac{1}{1 + r_{1}}C_{1}\right]$$ 

The houshold now chooses sequences of consumption and labor (i.e., representative household chooses $C_{0}, C_{1}, l_{0}, l_{1}$). The FOC along with the budget constraint imply a system of 5 equations in the 5 unknowns $C_{0}, C_{1}, l_{0}, l_{1}, \lambda$ as follows:

$$\frac{1}{C_{0}} - \lambda = 0$$

$$\beta\frac{1}{C_{1}} - \lambda \frac{1}{1+r_{1}}= 0$$

$$-\frac{b}{1 - l_{0}} + \lambda w_{0} = 0$$

$$-\beta\frac{b}{1 - l_{1}} + \lambda \frac{1}{1 + r_{1}}w_{1} = 0$$

$$C_{0} + \frac{1}{1 + r_{1}}C_{1} = w_{0}l_{0} + \frac{1}{1 + r_{1}}w_{1}l_{1}$$

This 5 equation system can be reduced (by eliminating the Lagrange multiplier $\lambda$) to a linear system of 4 equations in 4 unknowns:

$$\begin{vmatrix} b & \ 0 & \ w_{0} & 0 \\\ \beta(1 + r_{1}) & -1 & 0 & 0 \\\ 0 & \ b & 0 & w_{1} \\\ 1 & \frac{1}{1 + r_{1}} & -w_{0} & -\frac{1}{1+r_{1}}w_{1} \end{vmatrix} \begin{vmatrix}C_{0} \\\ C_{1} \\\ l_{0} \\\ l_{1}\end{vmatrix} = \begin{vmatrix} w_{0} \\\ 0 \\\ w_{1} \\\ 0\end{vmatrix}$$

The above system can be solved in closde form using some method like Cramer's rule/substitution etc to yield the following optimal sequences/policies for consumption and labor supply:

$$C_{0} = \frac{1}{(1 + b)(1 + \beta)}\left(w_{0} + \frac{1}{1 + r_{1}}w_{1}\right) $$

$$C_{1} = \left(\frac{1 + r_{1}}{1 + b}\right)\left(\frac{\beta}{1 + \beta}\right)\left(w_{0} + \frac{1}{1 + r_{1}}w_{1}\right) $$

$$l_{0} = 1 - \left(\frac{b}{w_{0}}\right)\left(\frac{1}{(1 + b)(1 + \beta)}\right)\left(w_{0} + \frac{1}{1 + r_{1}}w_{1}\right) $$

$$l_{1} = 1 - \left(\frac{b\beta(1+r_{1})}{w_{1}}\right) \left(\frac{1}{(1 + b)(1 + \beta)}\right)\left(w_{0} + \frac{1}{1 + r_{1}}w_{1}\right) $$

Several important points to note about the above optimal consumption and labor supply policies:

Household's lifetime net worth, $w_{0} + \frac{1}{1 + r_{1}}w_{1}$, is the present discounted value of its labor endowment.
Household's lifetime net worth depends on the wages in BOTH periods and future interest rate.  In hints at the more general result that, if the household has an infinite time horizon, lifetime net worh depends on the entire future path of wages and interest rates!
In each period, houshold's consume a fraction of their lifetime net worth. Although the fraction changes in this simple two period model, if the household has an infinite horizon, the fraction of lifetime networth consumed each period will be fixed and equal to 

$$\frac{1}{(1 + b)(1 + \beta + \beta^2 + \dots)}=\frac{1 - \beta}{1 + b}$$

From the policy function for $l_{0}$, one can show that in order for the labor supply in period $t=0$ to be non-negative (which it must!), the following inequality must hold: 

$$\left(\frac{1}{1 + r_{1}}\right)\left(\frac{w_{1}}{w_{0}}\right) \lt \frac{(1 + b)(1 + \beta)}{b} - 1$$

From the policy function for $l_{1}$, in order for the labor supply in period $t=1$ to be non-negative (which it must!), the following inequality must hold: 

$$(1 + r_{1})\left(\frac{w_{0}}{w_{1}}\right) \lt \left(\frac{1 + b}{b}\right)\left(\frac{1 + \beta}{\beta}\right) - 1$$

#Solving the RBC Model

The remainder of the lab focuses on solving and analyzing the RBC model using a numerical method from Uhlig (1997).  The method is a baby version of a very general technique for solving such models implemented in programs such as Dynare (a MatLab/Octave based plaform) and PyMacLab (a Python based tool that I am developing with Eric Scheffel).  The method can be broken down into the following steps:

1. Find the necessary equations characterizing the equilibrium ie constraints, FOCs, etc; define parameters.
2. Find the steady state(s).
3. Log-linearize the necessary equations characterizing the equilibrium of the system to make the equations approximately linear in the log-deviations from the steady state.
4. Solve for the recursive equilibrium law of motion via the method of undetermined coefficients.
5. Analyze the solution via impulse response analysis, and analyze the second order properties pf the model (possibly after applying the Hodrick-Prescott (HP) filter).

$$0= w_{t} - (1-\alpha)k_{t}^{\alpha} l_{t}^{-\alpha}$$

$$0= r_{t} - \alpha k_{t}^{\alpha - 1}l_{t}^{1 - \alpha} - \delta$$

```python
# my solutions
def F1(k, i):
    """
    
    Implicit function relating capital per effective worker and investment per effective
    person.
    
    """
    return np.exp(n + g) * k - ((1 - delta) * k + i)
    
def F2(y, c, i, gov):
    """
    
    Implicit function function relating output to consumption, investment, and government 
    purchases all per effective worker (you should recognize this as ye ole national 
    accounting identity!).
    
    """
    return y - (c + i + gov) 
     
def F3(y, k, l):
    """
    
    Implicit function describing the firms' production technology.
    
    """ 
    return y - k**alpha * l**(1 - alpha)

def F4(c, R):
    """
    
    The consumption Euler equation in implicit form.  Note that the expectations operator
    has been dropped because we are interested in the deterministic steady-state!
    
    """
    return (1 / c) - np.exp(-(rho + g)) * (1 / c) * R
    
def F5(c, l, w):
    """
    
    The intra-temporal labor supply FOC in implicit form.
    
    """
    return b * c - (1 - l) * w
                
def F6(k, l, w):
    """
    
    Implicit function relating the real wage (per effective worker!) to capital per 
    effective person.
    
    """
    return w - (1 - alpha) * k**alpha * l**(-alpha)
    
def F7(k, l, R):
    """
    
    Implicit function relating the real interest rate to capital per effective
    person.
    
    """
    return R - alpha * k**(alpha - 1) * l**(1 - alpha) - (1 - delta)
```

The standard approach in the literature to solving models of this type is to linearize (or more recently to take a higher order Taylor approximation, 2nd, or at most 3rd order) or log-linearize the model equations around the deterministic steady state. We are going to follow Romer (and many, many others!) and log-linearize our model around the deterministic steady state. The above implicit functions describe the behavior of the economy in the deterministic steady-state. Note that there are exactly 7 equations to go with 7 unknowns (i.e., y, k, c, i, l, w, r). There is actually an eighth "unknown", government purchases per effective worker, which is exogenous and will be treated as a free parameter to be set via calibration.

In steady-state, these equations will satisfy:

$$ F_{1}(k, i) = 0 $$

$$ F_{2}(y, c, i, gov) = 0 $$

$$ F_{3}(y, k, l,w,r) = 0 $$

$$ F_{4}(c, r) = 0 $$

$$ F_{5}(c, l, w) = 0 $$

$$ F_{6}(k,l,w) = 0 $$

$$ F_{7}(k, l, r) = 0 $$

Thus the problem of finding the steady-state values for our system of equations is "simply" a problem of finding the root of a 7-equation non-linear system! For a given level of government purchases, the above equations form a system of 7 non-linear equations in 7 unknowns: y, k, c, i, l, w, and r which we need to solve in order to obtain the steady-state values.

```python
# Define the function to be optimized
def rbcSteadyState(X, gov):
    """
    
    X is a vector containing the endogenous variables, X = [k, y, c, i, l, w, r]
    
    """
    out = [F1(X[0], X[3])]
    out.append(F2(X[1], X[2], X[3], gov))
    out.append(F3(X[1], X[0], X[4]))
    out.append(F4(X[2], X[6]))
    out.append(F5(X[2], X[4], X[5]))
    out.append(F6(X[0], X[4], X[5]))
    out.append(F7(X[0], X[4], X[6]))
    return out
```

##Task 5: Solving the model via log-linear approximation

The Python code in the cell below solves the RBC model using a method from Uhlig (1997).  The method is a baby version of a very general technique for solving such models implemented in programs such as Dynare (a MatLab/Octave based plaform) and PyMacLab (a Python based tool that I am developing with Eric Scheffel).  The method can be broken down into the following steps:

1. Find the necessary equations characterizing the equilibrium ie constraints, FOCs, etc. Done!
2. Pick parameters and find the steady state(s). Done!
3. Log-linearize the necessary equations characterizing the equilibrium of the system to make the equations approximately linear in the log-deviations from the steady state.
4. Solve for the recursive equilibrium law of motion via the method of undetermined coefficients.
5. Analyze the solution via impulse response analysis, and analyze the second order properties of the model (possibly after applying the Hodrick-Prescott (HP) filter).

Since we have already completed steps 1 and 2, we will begin with step 3: log-linearization.  The principle of log-linearization is to use a Taylor approximation around the steady state to replace all equations by approximations which are linear functions in the log deviations of the variables.  Log-linearization will allow us to linearize the our non-linear system of equations without having to explicitly taking any derivatives. For the details of how to apply this technique see section 3 of Uhlig (1997).

The code takes as input a list of coefficient matrices from the log-linearized model which approximates the  behaviour of our non-linear system in the neighborhood of the deterministic steady state, and then computes matrices describing the recursive equilibrium law of motion for the model. Specifically the code will compute matrices $P, Q, R, S$ satisfying the following:

$$\begin{align}\tilde{K_{t}} =& P \tilde{K_{t-1}} + Q \begin{vmatrix} \tilde{A_{t}} \\\ \tilde{G_{t}} \end{vmatrix} \\\ \begin{vmatrix} \tilde{Y_{t}} \\\ \tilde{C_{t}} \\\ \tilde{I_{t}} \\\ \tilde{L_{t}} \\\ \tilde{W_{t}} \\\ \tilde{R_{t}}\end{vmatrix}=& R \tilde{K_{t-1}} + S \begin{vmatrix} \tilde{A_{t}} \\\ \tilde{G_{t}} \end{vmatrix}\end{align}$$

The above equations express all of the endogenous variables of our model as function of the three state variables $K, \tilde{A}, \tilde{G}$. In our simple model, with only a single endogenous state variable (i.e., $K$), the matrix $P$ will actually be a scalar.  The matrix $Q$ will be (1, 2) because we have 2 exogenous stochastic processes describing technology shocks (i.e., $\tilde{A}$) and government spending shocks (i.e., $\tilde{G}$). The matrix $R$ will be (6, 6) because there are 6 endogenous, non-state variables in the model. Finally, the matrix $S$ will be (6, 2).
