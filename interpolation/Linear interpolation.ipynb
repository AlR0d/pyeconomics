```python
import numpy as np
from scipy import interpolate

from pyeconomics.interpolation import pwl_interp
```

### Basic 1D example:

```python
# knots and values
x = np.linspace(0, 10, 10)
y = np.cos(-x**2 / 8.0)

# points at which to evalute the function
xnew = np.linspace(0, 10, 40)
```

```python
# interpolate using SciPy's linear interpolation routine
f  = interpolate.interp1d(x, y, kind='linear')

# interpolate using SciPy's Univariate spline
f2 = interpolate.InterpolatedUnivariateSpline(x, y, k=1)

# interpolate using pwl_interp
f3 = pwl_interp.pwl_interp_1d(x, y, xnew)
```

```python
# plot the results...they should be identical
plt.plot(x, y, 'o', xnew, f(xnew),'-', xnew, f2(xnew), '--', xnew, f3, '.')
plt.legend(['data', 'interp1d', 'spline', 'pwl_interp'], loc='best')
plt.xlabel('x')
plt.show()
```

### Speed comparisons:

### Extrapolation:

All three methods can be made to exatrapolate to varying degrees of success.

```python
# points at which to evalute the function
xnew = np.linspace(-0.5, 10.5, 40)
```

```python
# interpolate using SciPy's linear interpolation routine
f  = interpolate.interp1d(x, y, kind='linear', bounds_error=False, fill_value=np.mean(y))

# interpolate using SciPy's Univariate spline
f2 = interpolate.InterpolatedUnivariateSpline(x, y, k=1, bbox=[-0.5, 10.5])

# interpolate using pwl_interp
f3 = pwl_interp.pwl_interp_1d(x, y, xnew)
```

```python
# plot the results...they should be identical
plt.plot(x, y, 'o', xnew, f(xnew),'-', xnew, f2(xnew), '--', xnew, f3, '.')
plt.legend(['data', 'interp1d', 'spline', 'pwl_interp'], loc='best')
plt.xlabel('x')
plt.show()
```

## Simple 2D example

```python
# Suppose we want to interpolate the 2-D function
def func(x, y):
    return x * (1 - x) * np.cos(4 * np.pi * x) * np.sin(4 * np.pi * y**2)**2

# on a grid in [0, 1]x[0, 1]
x, y = np.linspace(0, 1, 20), np.linspace(0, 1, 20)
grid_x, grid_y = np.meshgrid(x, y)

# but we only know its values at the following points:
values = func(grid_x, grid_y)

# values we wish to interpolate
xnew, ynew = np.linspace(0, 1, 50), np.linspace(0, 1, 50)
grid_xnew, grid_ynew = np.meshgrid(xnew, ynew)
```

```python
# 2D interpolation on a rectangular mesh using SciPy's routine
surf  = interpolate.RectBivariateSpline(x, y, values, kx=1, ky=1, s=0)

# dimensions are reversed!
surf2 = pwl_interp.pwl_interp_2d(x, y, values.T, grid_xnew.flatten(), grid_ynew.flatten()) 
```

```python
plt.figure(figsize=(12,8))
plt.subplot(131)
plt.imshow(func(grid_x, grid_y), extent=(0,1,0,1), origin='lower')
plt.title('Original')

plt.subplot(132)
plt.imshow(surf(xnew, ynew), extent=(0,1,0,1), origin='lower')
plt.title('RectBivariateSpline')

plt.subplot(133)
plt.imshow(surf2.reshape(grid_xnew.shape), extent=(0,1,0,1), origin='lower')
plt.title('pwl_interp_2d')
plt.show
```

```python
%timeit interpolate.RectBivariateSpline(x, y, values, kx=1, ky=1, s=0)
```

```python
%timeit pwl_interp.pwl_interp_2d(x, y, values, grid_xnew.flatten(), grid_ynew.flatten()) 
```
